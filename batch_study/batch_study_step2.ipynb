```python
# Import libraries
import pandas as pd
import numpy as np
from reComBat import reComBat
import time

# Set up directories and variables
folder = "batch_study"
exp_folder = folder + "/experiment_24/"

covmat = pd.read_csv(f"{exp_folder}covmat.csv", index_col=0)
covmat.reset_index(inplace=True, drop=True)
covmat = covmat.astype(str)

recombat_csvs = []
time_recombat = []
```

```python

for iter in range(1,6):
    print("Iteration",iter)
    batch_df = pd.read_csv(f"{exp_folder}iter{iter}_batch_df_BS.csv", index_col=0)
    
    # metadata
    metadata = pd.read_csv(f"{exp_folder}iter{iter}_metadata_BS.csv", index_col=0)
    batches_series = metadata["batch"]
    batches = batches_series.to_numpy(dtype="<U6")

    # reComBat calc
    count_batch_transformed = pd.read_csv(f"{exp_folder}iter{iter}_countmat_batch_transformed_BS.csv", index_col=0)
    start = time.time()
    recombat_model = reComBat(model='ridge', parametric=True, config={'alpha': 1e-9})
    recombat_df = recombat_model.fit_transform(count_batch_transformed.T, batches_series, X=covmat)
    end = time.time()
    recombat_df.T.to_csv(f"{exp_folder}iter{iter}_recombat_df_BS.csv")

    time_recombat.append((end-start)/60)
print(np.mean(time_recombat))
```
